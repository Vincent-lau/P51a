# 2. Examples of graphs

These are examples using matplotlib.pyplot but you are free to use something else.

```python
# Quick example to start
from random import randint
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# plot one line
x1 = np.arange(0,100,10)
y1 = sorted([randint(0,100) for i in range(10)])
plt.plot(x1,y1,label="line1")

# plot another line
x2 = np.arange(0,100,1)
y2 = sorted([randint(0,100) for i in range(100)])
plt.plot(x2,y2,label="line2")

# show on the same figure
plt.legend()
plt.title('Title')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.yscale('log')
plt.show()
```

### 2.1 CDF

```python
ran = [randint(0,100)*randint(0,100) for i in range(500)] # a list of values
values, base = np.histogram(ran, bins=1000)
cumulative = np.cumsum(values/500.0) # divide by number of values
plt.plot(base[:-1], cumulative)
plt.show()
```

### 2.2 Error bars
[Docs](https://matplotlib.org/devdocs/api/_as_gen/matplotlib.pyplot.errorbar.html#matplotlib.pyplot.errorbar)    
NB For yerr, if a sequence of shape 2xN, errorbars are drawn at -row1 and +row2 relative to the data. Where N is length of x or y array.

```python
xs = [1,2,3,4,5]
ys = sorted([randint(0,10) for i in range(5)])
yerr = [[randint(0,10)/10.0 for i in range(5)],[randint(0,10)/10.0 for i in range(5)]]
plt.errorbar(xs,ys,yerr=yerr)
plt.show()
```

```python
# To plot error bars over an existing plot
plt.errorbar(xs,ys,yerr=yerr,linestyle='none')
plt.show()
```

### 2.3 Step graph
[Docs](https://matplotlib.org/devdocs/api/_as_gen/matplotlib.pyplot.step.html)  
NB If where = 'post', the interval from x[i] to x[i+1] has level y[i]. See from the demo that the last value is a vertical line. So if you want to not have that, you must append the last value to the list, so that the line becomes a dot.  

```python
xs = np.arange(0,11,1)
ys = [0+randint(0,10) for i in range(10)]
ys.append(ys[-1])
plt.step(np.arange(0,11,1), ys ,where="post")
plt.show()
```

### 4. Histogram

```python
data = []
for i in range(1000):
    data.append(25)
for i in range(20):
    data.append(20)
for i in range(50):
    data.append(22)
plt.hist(data,log=True,bins=50)
plt.show()
```
